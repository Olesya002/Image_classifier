```python
# Вычислим среднеквадратичную контрастность изображения
def calculate_image_contrast(pic):
  tensor_pic = torch.tensor(imread(pic, as_gray=True))
  mean_brightness = torch.sum(tensor_pic)/(tensor_pic.shape[0]*tensor_pic.shape[1])
  return (torch.sum((tensor_pic - mean_brightness) ** 2) / (tensor_pic.shape[0]*tensor_pic.shape[1])) ** (1/2)
```

```python
def picture_partition(pic, n):
  tens_pic = torch.tensor(imread(pic, as_gray=False), dtype=float)
  n1, n2 = int(n ** (1/2)), n // int(n ** (1/2))
  width, height = tens_pic.shape[1], tens_pic.shape[0]
  if n1 >= n2:
    if width >= height:
      step1, step2 = width // n1, height // n2
    else:
      step1, step2 = width // n2, height // n1
  else:
    if width >= height:
      step1, step2 = width // n2, height // n1
    else:
      step1, step2 = width // n1, height // n2
  tens_parts = []
  for i in range(0, height, step2):
    for j in range(0, width, step1):
      tens_parts.append(tens_pic.narrow(0, i, step2).narrow(1, j, step1))
  return tens_parts
```

```python
def calculate_image_contrast_1(pic):  
  temp = torch.tensor(imread(pic, as_gray=False))
  temp = temp / 255
  id1 = temp > 0.03928
  id2 = temp <= 0.03928
  temp[id1] = (((temp[id1] + 0.055)/ 1.055) ** 2.4)
  temp[id2] = temp[id2] / 12.92
  R = temp.permute(2,0,1)[0] * 0.2126
  G = temp.permute(2,0,1)[1] * 0.7152
  B = temp.permute(2,0,1)[2] * 0.0722
  gray = R + B + G
  return ((torch.max(gray) + 0.05)/(torch.min(gray) + 0.05))
```

```python
def calculate_sharpness(pic):
  im = Image.open(pic).convert('L') # тоже возвращает в оттенках серого
  array = np.asarray(im, dtype=np.int32)
  gy, gx = np.gradient(array)
  gnorm = np.sqrt(gx**2 + gy**2)
  return np.average(gnorm)
```

```python
def calculate_corners(pic):
  gray = np.array(Image.open(pic).convert('L'))
  dst = cv2.cornerHarris(gray, 3, 5, 0.1)
  corners = dst > 0.05 * dst.max()
  return np.sum(corners)
```

Признак Total Variance (общая дисперсия) - это мера изменчивости или разброса значений величины в наборе данных. Для изображения это может быть выражено как сумма квадратов отклонений каждого пикселя от среднего значения по всем пикселям изображения. Чем больше значение Total Variance, тем больше разнообразие яркости или цветовых значений в изображении.
Такой признак позволяет оценить насколько "размытыми" или разнообразными являются данные на изображении, и может быть полезным при задачах анализа изображений, сегментации или классификации.
